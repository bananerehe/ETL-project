```python
import pandas as pd
from sqlalchemy import create_engine
import psycopg2
```

```python
happiness_file = "data/world-happiness-report.csv"
happiness_df = pd.read_csv(happiness_file)
happiness_df=happiness_df.rename(columns={'Country name':'Entity'})
happiness_df['key']=happiness_df['Entity']+'_'+happiness_df['year'].astype('str')
happiness_df.head()
```

```python
happiness_df.columns
```

```python
veggies_file = "data/vegetable-consumption-per-capita.csv"
veggies_df = pd.read_csv(veggies_file)
veggies_df['key']=veggies_df['Entity']+'_'+veggies_df['Year'].astype('str')
# Rename the column headers
veggies_transform = veggies_df.rename(columns={"Vegetables - Food supply quantity (kg/capita/yr) (FAO, 2020)": "Veggies kg/capita",
    })

#veggies_transformed = veggies_transform ["Entity","Year", "Veggies (kg/capita)","key"]
veggies_transform.head()
```

```python
fruits_file = "data/fruit-consumption-per-capita.csv"
fruits_df = pd.read_csv(fruits_file)
fruits_df['key']=fruits_df['Entity']+'_'+fruits_df['Year'].astype('str')
# Rename the column headers
fruits_df = fruits_df.rename(columns={"Fruits - Excluding Wine - Food supply quantity (kg/capita/yr) (FAO, 2020)": "Fruits kg/capita",
    })
fruits_df.head()
```

```python
# Create a filtered dataframe from specific columns
# Dropped Social support, Freedom to make life choices, Generosity, Perceptions of corruption, Positive affect, Negative affect
happiness_cols = ["Entity","year","Life Ladder","Log GDP per capita","Healthy life expectancy at birth",'key']
happiness_transformed= happiness_df[happiness_cols].copy()
happiness_transformed = happiness_transformed.rename(columns={"year": "Year",
    })
happiness_transformed.head()
```

```python
rds_connection_string = "postgres:postgres@localhost:5432/happiness_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
engine.table_names()
```

```python
# Load Happiness data
happiness_transformed.to_sql("happiness_df", con=engine, if_exists="append", index=False)
```

```python
# Load Fruits data
fruits_df.to_sql("fruits_df", con=engine, if_exists="append", index=False)
```

```python
# Load Veggie data
veggies_transform.to_sql("veggies_df", con=engine, if_exists="append", index=False)
```
